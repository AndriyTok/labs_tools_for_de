### Глибоке навчання.

Глибоке навчання — це підхід до машинного навчання, який використовує багатошарові нейронні мережі для вирішення складних завдань.

Нейронні мережі — це моделі машинного навчання, які імітують структуру та функції людського мозку. Вони складаються з нейронів, які пов’язані між собою. Нейрони можуть обробляти інформацію та навчатися на основі даних. 

### Історія

* 1940-ві: розробка перших нейронних мереж;
* 1950 - 2000:  розробка нових методів навчання та архітектур нейронних мереж;
* 2000 - 2010: розвиток обчислювальних ресурсів дозволив ефективно навчати нейронні мережі з великою кількістю шарів;
* 2010 - наші дні: активний розвиток глибокого навчання.

### Переваги та недоліки.

Переваги:
* Розв’язок завдань, складних для звичайного машинного навчання;
* Робота з великими обсягами даних;
* Можливість адаптації до нових завдань без необхідності перенавчання.

Недоліки: 
* Вимагає великих обсягів даних для навчання;
* Складне для розуміння та інтерпретації;
* Схильне до перенавчання.

### Застосування.

* Розпізнавання образів;
* Розпізнавання мови;
* Машинний переклад;
* Комп’ютерна візуалізація;
* Обробка природної мови;
* Генеративне навчання.

### Перспективи.
* Розвиток нових архітектур нейронних мереж;
* Вдосконалення методів навчання нейронних мереж;
* Розвиток нових застосувань глибокого навчання.


### Нейронні мережі.

Нейронні мережа - це послідовність алгоритмів, які намагаються розпізнати зв'язки між даними через процес, який мімікрує спосіб за яким працює людський мозок.
Проста нейрона мережа має приблизно такий вигляд: <br>

* Іnput layer - вхідний шар. Через цей шар дані передаються всередину ситеми для майбутньої обробки нейронами.
* Hidden layer - прихований шар. Цей шар обробляє дані та передає їх на наступнний шар.
* Output layer - вихідний шар. Цей шар отримує інформацію від попередніх шарів та робить пердбачення.

Вузли такої мережі називаються нейронами.
<br> Модель, яку показану на зображенні вище, можна назвати поверхневою (shalow), оскільки у моделі є тільки один прихований рівень.
<br> 
<br> На зображені вище показна модель, яка має три приховані рівні. У цьому випадку говорять про глибоке навчання, оскільки рівнів є більше ніж один. Окрім того, тут мережа є повністю з'єднанна, оскільки всі нейрони мають зв'язки один між одним.

### Нейрон.


Нейрон - це основний будівельний блок штучних нейронних мереж.<br>

Тут:
* $x_1 - x_m$ - вхідні дані. Вхідні дані можуть бути як нормалізовані, так і не нормалізовані;
* $w_k1 - w_km$ - ваги. Ваги визначають наскільки то чи інше вхідне значення впливає на кінцевий результат;
* $b_k$ - зміщення. Це є константне значення, яке використовується для впливу на вихідні значення. Воно дозволяє змістити значення функції активації в позитивну або негативну сторону.
<br>
Вихідні значення можуть бути різного типу: числовими, бінарними або категоріальними. У випадку бінарної або категоріальної змінної виходом буде не один нейрон, а набір, кожен з яких буде відповідати одній з категорій.
Всередені нейрона всі дані, які подаються на вхід множаться на значення ваг, додаються і після застосовується функція активації. Ця функція визначає, чи потрібно передавати дані у наступні нейрони.

### Процес навчання.
Одношарова feed forfard neural network називається персептроном. Нехай персептрон опрацював певні дані та дав певне вихідне значення. Це значення порівнюється з значенням, яке було отримане у реальності. Обраховується функція втрат. Ця функція говорить, наскільки велика є помилка у зробленому прогнозі. Завдання перед розробниками - мінімізувати функцію втрат. Після порівняння та обрахунку значення функції, ця інформація передається назад у мережу, і з мережі вона передається до ваг, які після цього оновлюються. Цей процес називається back propagation.<br>
Цей процес відбувається з одним спостереженням. І це повторюється доти, доки функція втрат не поверне значення, менше за певне значення.<br>
Однією епохою навчання називають процес проходження через весь набір даних.

### Градієнтний спуск.
Постає питання, яким чином мінімізовувати функцію втрат. Простий перебір значень (brute force). Такий підхід є надзвичайно простим і зрозумілим. Але він може працювати у випадках з дуже малими мережами. З більшими мережами, коли збільшується кількість зв'яків між нейронами, з'являється таке поняття як "Прокляття вимірності (Curse of Dimensionality)". <br>
Розглянемо приклад:
Нехай є проста мережа, у якої є 25 синапсисів (зв'язків між нейронами). Для простого перебору 1000 значень з мережею у якій 25 синапсисів отримуємо:<br>
1000 * 1000 * 1000 * 1000 * ... * 1000 = $1000^{25}$ = $10^{75}$ комбінацій.  Це число комбінацій, яке потрібно протестувати методом простого підбору.<br>
Для вирішення цієї проблеми використовується так званий "градієнтний спуск".
<br>
 <br>
У початковій точці розглядається кут нахтлу функції втрат. У випадку додатнього нахилу, функція рухається по схилу, у випадку від'ємного - під гору. Цей процес повторюється доти, доки не знайдено оптимальне значення.
Це приклад найпростішого градієнтного спуску, який може знайти оптимальне значення для функцій, які мають лише один глобальний мінімум. Якщо функція має декілька локальних мінімумів, простий градієнтий спуск не завжди зможе знайти найоптиальніше значення. 
У цьому випадку використовуються складніші меоди, як, наприклад стохастичний градінтний спуск <a href=''>SGDM</a>.

Метод градієнтного спуску є ефективним для навчання нейронних мереж, оскільки він дозволяє мережі рухатися в напрямку, у якому функція помилки зменшується найбільше. Це дозволяє мережі швидко знаходити оптимальне рішення. 
Ось кілька переваг методу градієнтного спуску:
* Ефективність;
* Простота; 
* Універсальність.

Однак, метод градієнтного спуску має і деякі недоліки:
* Можливість застрягти в місцевому мінімуму;
* Нестабільність.

### Вибуховий та зникаючий градієнти.
__Вибухові градієнти__ виникають у нейронних мережах, коли значення градієнта функції втрат стають дуже великими під час зворотного поширення. Це може призвести до нестабільності навчання та ускладнити зближення моделі з оптимальним рішенням.

__Зникаючі градієнти__ - це протилежна проблема, коли значення градієнта стають дуже маленькими під час зворотного поширення. Це може призвести до того, що модель не зможе навчатися, оскільки оновлення ваги стають незначними.

Причини виникнення:
* Глибина нейронних мереж;
* Нелінійні функції активації;
* Неправильна ініціалізація ваги.
