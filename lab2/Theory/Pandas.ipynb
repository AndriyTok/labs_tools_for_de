# Pandas

Pandas - це потужний пакет Python з відкритим кодом, який використовується для аналізу та обробки даних. Він надає зручні та інтуїтивно зрозумілі інструменти для роботи з табличними даними, часовими рядами та категоріальними змінними.

Основні структури даних Pandas:

Series: одновимірний масив даних, подібний до списку Python, але з додатковими можливостями для індексування та вибірки.
DataFrame: двовимірна таблиця даних, подібна до електронної таблиці, з рядками, стовпцями та індексами.

Для вибору рядків використовуються властивості [loc](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.loc.html) та [iloc](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iloc.html).

`loc` доступається до групи рядків та колонок по мітці (мітках) або масиву булевих значень.

```python
loc_df = pd.DataFrame([[1,2], [3,4], [5, 6]],
                      index=['first', 'second', 'third'],
                      columns=['first_column', 'second_column'])
loc_df
```

```python
loc_df.loc['first']
```

Також можна вибирати по критерію з використанням булевих змінних. Наприклад, для вибору всіх записів, де країна не є None:

```python
df.loc[df['Country'] != 'None']
```

`iloc` в свою чергу працює лише з цілочисельними індексами.

```python
df.iloc[2, 3]
```

У прикладі зверху 2 - це індекс рядка а 3 індекс стовпця.

### Slicing (нарізка).
Нарізка дозволяє вибирати певні набори даних з масивів, а не тільки один елемент.
При використанні нарізки перше число включається у результат, коли останнє не враховується.

Також, допускається не вказування чисел при нарізці. Тоді у нарізці будуть використовуватись всі елементи. 

```python
df.iloc[2:5, :]
```

### Видалення.

Для видалення рядочків та стовпців використовується метод drop. У ньому важливо вказати два параметри: `lables` який вказує, який саме масив треба видалити та `axis`, який вказує по якій осі шукати масив, який треба видалити. `axis = 0` буде шукати масив по рядках, а `axis = 1` по стовпчиках.

### Вибір з умовою.

З використанням базових умов:

```python
df[df['HighestRanking'] == 1]
```

Використовуючи `loc`:

```python
df.loc[df['HighestRanking'] == 1]
```

Використовуючи метод [isin()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.isin.html):

```python
options = [1,2,3]
df[df["HighestRanking"].isin(options)]
```

Також можна обирати елементи, які не задовільняють умови `options`:

```python
options = [1,2,3]
df[~df["HighestRanking"].isin(options)]
```

Базові умови та `isin()` можуть бути скомбіновані:

```python
options = [1,2,3]
df[(df['Continent'] == 'Europe') & df['HighestRanking'].isin(options)]
```

### [map()](https://pandas.pydata.org/docs/dev/reference/api/pandas.DataFrame.map.html)

Застосовує функцію до кожного елемента DataFrame.

```python
sample['Continent'].map('The continent is {}'.format)
```

Співставлення може відбуватись з використанням словника:

```python
sample['Continent'].map({'None' : 'Continent not found'})
```

У прикладі зверху для елементів, для яких не вказані елементи у словнику виводиться NaN. Щоб вивести нормальне значення можна використати два підходи:
1. задати у словнику значення для усіх можливих елементів;
1. задати лямбда функцію:

```python
sample['Continent'].map(lambda x: 'Continent not found' if x == 'None' else x)
```

### Групування та сортування.

__Сортувати__ дані можна у зростаючому та спадаючому порядках. За це відповідає параметр `ascending` методу [sort_values()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html). Для того щоб вказати по якому стовпчику сортувати, використовується параметр `by`, який приймає список стовпчиків, по якому сортувати дані.

Аналогічно можна сортувати по рядочках, але для цього використовується метод [sort_index()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sort_index.html).

```python
df.sort_values(by = ['RegisterDate'], ascending = False) #-> by default ascending
```

__Групування__ даних може відбуватись по одному або більше стовпчиках. Для групування треба виликати метод [groupby()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html) та вказати у параметрах які стовці використати для групування.

```python
df.groupby('Continent').count()
```

```python

df.groupby(['Continent', 'Country', 'HighestRanking']).count()
```

Агрегування можна застосовувати як до всього DataFrame, так і до окремих стовпців, якщо використати словник.

```python
df.agg({'Points': 'sum', 'HighestRanking': 'min'})
```

## Об'єднання даних.

### [join()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.join.html)

Додає колонки іншого датафрейму.

```python
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': ['a', 'b', 'c']}, index=['x', 'y', 'z'])
df2 = pd.DataFrame({'C': [4, 5, 6]}, index=['y', 'z', 'x'])

df1.join(df2)
```

### [merge()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html)
Об'єднує DataFrame або поіменовані Series у вигляді join-ів з баз даних.

```python
df1 = pd.DataFrame({'key': ['a', 'b', 'c'], 'value1': [1, 2, 3]})
df2 = pd.DataFrame({'key': ['b', 'c', 'd'], 'value2': [4, 5, 6]})

pd.merge(df1, df2, on='key')
```

### [concat()](https://pandas.pydata.org/docs/reference/api/pandas.concat.html)
Об'єднує два датафрейми вздовж обраної осі.

```python
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], ignore_index=True)
```

## Перетворення даних.

### [apply()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html)
Дозволяє застосувати обрану функцію до даних вздовж певної осі.

```python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

def my_function(row):
    return row['A'] + row['B']

df.apply(my_function, axis=1)
```
